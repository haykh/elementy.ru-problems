```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rc
```

```python
def k_form(f):
  return (0.8177399999999999*(1 - f) + 3.256695*f)/(2.969*(1 - f) + 1.324*f)

with plt.xkcd():
  fig, ax = plt.subplots(figsize=(8, 6), dpi=100)
  xs = np.linspace(0, 1, 100)
  ys = k_form(xs)
  ax.plot(xs, ys)
  ax.set_title(r'цепная реакция для быстрых нейтронов')
  ax.axhline(1, xmax=0.6, c='gray', ls=':')
  ax.axvline(0.53, ymax=0.4, c='gray', ls=':')
  ax.set_xlabel(r'$f$ (доля ${}^{235}_{~~94}\mathrm{U}$)')
  ax.set_ylabel(r'$k$ (среднее число вторичных нейтронов)')
  ax.text(0, 1, 'порог для цепной реакции', ha='left', va='bottom', c='gray')
  ax.text(0.55, 0.5, 'минимальный уровень\nобогащения', ha='left', va='center', c='gray')
```
