```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import matplotlib
```

```python
K = 100;
x = 3e-10;
alpha = 0.02;
k = 5000;
q = 1e4;
T = 100;

def ode(y, t, K, x, alpha, k, q, T):
  N, n = y
  dydt = [K - x * N * n, alpha * K * k + q * x * N * n - n / T]
  return dydt

N = np.logspace(3, 5, 100)
n = np.logspace(4, 10, 100)

N, n = np.meshgrid(N, n)

dN, dn = ode((N, n), ..., K, x, alpha, k, q, T)

nc = T * (alpha * K * k + q * K)
Nc = K / (x * nc)
print (f'n = {nc:.2e}, N = {Nc:.2e}')

vN = dN / N
vn = dn / n
L = np.sqrt(vN**2 + vn**2)
```

```python
fig, ax = plt.subplots(figsize=(6, 4), dpi=200)

ax.pcolor(N, n, L, shading='auto', norm=matplotlib.colors.LogNorm(), cmap='binary_r')
skip=(slice(None,None,4),slice(None,None,4))
ax.quiver(N[skip], n[skip], (vN / L)[skip], (vn / L)[skip])

ax.fill_between(N[0], [1e7] * len(N), [1e10] * len(N), alpha=0.3, fc='red')
ax.text(3e4, 1.5e7, 'критическое\nколичество мусора', c='k', ha='left', va='bottom', fontsize=8,
        bbox={
          'fc': 'w', 
          'alpha': 0.8, 
          'lw': 0,
          'boxstyle': 'round',
        })

ax.scatter(Nc, nc, fc='w', s=50, marker='*', ec='k', linewidths=1)

ts = np.linspace(-19, 400, 250)
y0 = [5e3, 1e5]
cols = plt.cm.turbo((ts - ts[0]) / (ts[-1] - ts[0]))
sol = odeint(ode, y0, ts, args=(K, x, alpha, k, q, T))
ax.scatter(*(sol.T), c=cols)

ax2 = ax.inset_axes([0.5, 0.88, 0.49, 0.1])
ax2.pcolor(2020 + ts, [0, 1], [ts, ts], shading='auto', cmap='turbo')
ax2.pcolor(2020 + ts, [-0.5, 0], [0*ts, 0*ts], shading='auto', cmap='binary_r')
ax2.set_yticks([])
ax2.tick_params(axis="x", direction="in", pad=-8, labelsize=5, length=2, labelcolor='w', color='w')

ax.set_xlabel('количество спутников', fontsize=7)
ax.set_ylabel('количество "опасных" частиц мусора (массой > г)', fontsize=7)

ax.set_xscale('log'); ax.set_yscale('log');
```
