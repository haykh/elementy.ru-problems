```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
# binomial coefficient n k
def binomial(n, k):
    return np.math.factorial(n) / (np.math.factorial(k) * np.math.factorial(n-k))

def get4N(N):
    ks = np.arange(N + 1)
    ys = np.array([binomial(N, k) / 2**N for k in ks])
    return ks, ys

Ns = [10, 100, 1000]

fig = plt.figure(figsize=(8, 12), dpi=150)

gs = mpl.gridspec.GridSpec(3, 1)

for i, n in enumerate(Ns):
    ax = plt.subplot(gs[i, 0])
    a, b = get4N(n)
    ax.plot(a, b, '-o', color='k')
    ax.text(0.1, 0.9, transform=ax.transAxes, s=fr'$n={n}$', fontsize=16)
    ax.set_ylabel(r'вероятность')
    ax.set_xlabel(r'сумма, $m$')

plt.tight_layout()
```
