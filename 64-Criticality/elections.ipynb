```python
turnout = data.iloc[:, -6:-1].sum(axis=1) / data.iloc[:, 3]
```

```python
bins = [np.linspace(0, 100, 100), np.linspace(0, 100, 100)]
cnt, xe, ye = np.histogram2d(turnout * 100, putin * 100,
               bins=bins)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
ax.set_aspect(1)
ax.grid(False)

ax.hist2d(turnout * 100, putin * 100,
          bins=[np.linspace(0, 100, 100), np.linspace(0, 100, 100)],
          cmap='turbo', vmin=0, vmax=100)
ax.set_xlabel(r'Явка, \%')
ax.set_ylabel(r'Доля голосовавших за победителя, \%')
```
