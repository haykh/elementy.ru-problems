```python
%%manim EMWave -i --media_dir ./med --video_output_dir ./vid --vcodec prores --verbose

OSC_RUNTIME = 5
CAM_RUNTIME = 2
NTURNS = 2

class EMWave(ThreeDScene):
    CONFIG = {
        "camera_config": {
            "shading_factor": 1.0,
            "phi": 70 * DEGREES,
            "theta": 20 * DEGREES
        },
        "should_apply_shading": True,
        "color": False,
        "amplitude": 2
    }
    def construct(self):
        import numpy as np
        self.camera.set_frame_width(7)
        self.camera.set_frame_center([0, 1, 1])
        self.camera.reset_pixel_shape(700, 900)
        self.camera.resize_frame_shape()
        
        time_tracker = ValueTracker(0)
        phase_tracker = ValueTracker(0)
        anims = [ApplyMethod(time_tracker.increment_value, OSC_RUNTIME)]
        anims += [UpdateFromFunc(phase_tracker, 
                                 lambda x: x.set_value(2 * PI * ((time_tracker.get_value()) / (OSC_RUNTIME / NTURNS))))]
                
        def update_E(obj):
            shift_x = obj.get_start()[0]
            delay = shift_x
            length = self.amplitude * np.cos(phase_tracker.get_value() + delay)
            new_origin = obj.get_start()
            obj.put_start_and_end_on(new_origin, new_origin + np.array([0, length, 0]))
            obj.rotate(90 * DEGREES, DOWN)
        def update_B(obj):
            shift_x = obj.get_start()[0]
            delay = shift_x
            new_origin = obj.get_start()
            length = self.amplitude * np.cos(phase_tracker.get_value() + delay)
            old_length = obj.get_end()[2]
            obj.start = new_origin
            obj.end = new_origin + np.array([0, 0, length])
            obj.generate_points()
            obj.set_stroke_width_from_length()
            new_stroke_size = obj.get_stroke_width()/30
            tip = obj.get_tip()
            tip.set_height(new_stroke_size)
            tip.move_to(new_origin + np.array([0, 0, length + np.sign(length) * tip.get_height()/2]))
            tip_vertices_z = tip.get_vertices()[:, 2]
            looking_down = (tip_vertices_z[0] < tip_vertices_z[1])
            if ((looking_down and length > 0) or (not looking_down and length < 0)):
                tip.rotate(180*DEGREES, UP, about_point=obj.get_end())

        e_arrow = Arrow(start=ORIGIN, buff=0, tip_length=self.amplitude/10, stroke_color=BLUE)
        b_arrow = Arrow(start=ORIGIN, buff=0, tip_length=self.amplitude/10, stroke_color=RED)
        k_arrow = Arrow(start=[0,0,0], end=[2*self.amplitude,0,0], buff=0, tip_length=self.amplitude/10, stroke_color=WHITE)
        tip = b_arrow.get_tip()\
                .rotate(90*DEGREES, UP, about_point=b_arrow.get_end())
        
        self.add(e_arrow, b_arrow, k_arrow)
        
        Nghosts = 20
        for i in range(Nghosts):
            opacity = (Nghosts-1-i)/Nghosts
            shift_x = -(i+1)/2+1
            
            e_ghost = Arrow(start=ORIGIN, buff=0, tip_length=self.amplitude/10, stroke_color=BLUE_E, stroke_opacity=opacity)\
                        .move_to(np.array([shift_x, 0, 0])).set_shade_in_3d(True)
            tip = e_ghost.get_tip().set_opacity(opacity)
            self.add(e_ghost)
            anims += [UpdateFromFunc(e_ghost, update_E)]
            
            b_ghost = Arrow(start=ORIGIN, buff=0, tip_length=self.amplitude/10, stroke_color=RED_E, stroke_opacity=opacity)\
                        .move_to(np.array([shift_x, 0, 0])).set_shade_in_3d(True)
            tip = b_ghost.get_tip().set_opacity(opacity)\
                .rotate(90*DEGREES, UP, about_point=b_ghost.get_end())
            self.add(b_ghost)
            anims += [UpdateFromFunc(b_ghost, update_B)]
        
        anims += [UpdateFromFunc(e_arrow, update_E)]
        anims += [UpdateFromFunc(b_arrow, update_B)]
        
        text_e = TexMobject(r'\boldsymbol{E}', color=BLUE_C)\
                    .rotate(90*DEGREES, RIGHT).rotate(90*DEGREES, OUT)\
                    .move_to(2 * UP + 0.5 * IN)
        text_b = TexMobject(r'\boldsymbol{B}', color=RED_C)\
                    .rotate(90*DEGREES, RIGHT).rotate(90*DEGREES, OUT)\
                    .move_to(2 * OUT + 0.5 * DOWN)
        text_k = TexMobject(r'\boldsymbol{k}', color=WHITE)\
                    .rotate(90*DEGREES, RIGHT).rotate(90*DEGREES, OUT)\
                    .move_to(2.5 * RIGHT + 0.5 * UP)
        self.add(text_e, text_b, text_k)
        
        self.play(*anims, run_time=OSC_RUNTIME, rate_func=linear)
```

```python
%%manim EMWave2 -i --media_dir ./med --video_output_dir ./vid --vcodec prores --verbose

OSC_RUNTIME = 5
CAM_RUNTIME = 2
NTURNS = 2

class EMWave2(ThreeDScene):
    CONFIG = {
        "camera_config": {
            "shading_factor": 1.0,
            "phi": 70 * DEGREES,
            "theta": 20 * DEGREES
        },
        "should_apply_shading": True,
        "color": False,
        "amplitude": 2
    }
    def construct(self):
        import numpy as np
        self.camera.set_frame_width(7)
        self.camera.set_frame_center([0, 1, 1])
        self.camera.reset_pixel_shape(700, 900)
        self.camera.resize_frame_shape()
        
        d_angle = 2 * PI * NTURNS / (OSC_RUNTIME * self.camera.frame_rate)
        def update(obj):
            obj.rotate(d_angle, LEFT, about_point=obj.get_start())
        
        time_tracker = ValueTracker(0)
        phase_tracker = ValueTracker(0)
        anims = [ApplyMethod(time_tracker.increment_value, OSC_RUNTIME)]
        anims += [UpdateFromFunc(phase_tracker, 
                                 lambda x: x.set_value(2 * PI * ((time_tracker.get_value()) / (OSC_RUNTIME / NTURNS))))]
                


        e_arrow = Arrow(start=ORIGIN, end=ORIGIN+self.amplitude*UP, 
                        buff=0, tip_length=self.amplitude/10, stroke_color=BLUE)\
                    .rotate(90*DEGREES, UP)
        self.add(e_arrow)
        anims += [UpdateFromFunc(e_arrow, update)]
    
        Nghosts = 20
        for i in range(Nghosts):
            opacity = ((Nghosts-1-i)/Nghosts)**2
            shift_x = -(i+1)/2
                        
            origin = ORIGIN + np.array([shift_x,0,0])
            e_ghost = Arrow(start=origin, end=origin+self.amplitude*UP, buff=0, tip_length=self.amplitude/10, 
                            stroke_color=BLUE, stroke_opacity=opacity)\
                        .rotate(90*DEGREES, UP)\
                        .rotate(shift_x/2, LEFT, about_edge=DOWN)\
                        .set_shade_in_3d(True)
            tip = e_ghost.get_tip().set_opacity(opacity)
            self.add(e_ghost)
            anims += [UpdateFromFunc(e_ghost, update)]
    
        b_arrow = Arrow(start=ORIGIN, end=ORIGIN+self.amplitude*UP, 
                        buff=0, tip_length=self.amplitude/10, stroke_color=RED)\
                    .rotate(90*DEGREES, UP)\
                    .rotate(90*DEGREES, RIGHT, about_point=ORIGIN)
        self.add(b_arrow)
        anims += [UpdateFromFunc(b_arrow, update)]
        Nghosts = 20
        for i in range(Nghosts):
            opacity = ((Nghosts-1-i)/Nghosts)**2
            shift_x = -(i+1)/2
            
            origin = ORIGIN + np.array([shift_x,0,0])
            b_ghost = Arrow(start=origin, end=origin+self.amplitude*UP, buff=0, tip_length=self.amplitude/10, 
                            stroke_color=RED, stroke_opacity=opacity)\
                        .rotate(90*DEGREES, UP)\
                        .rotate(90*DEGREES, RIGHT, about_point=ORIGIN)\
                        .rotate(shift_x/2, LEFT, about_edge=IN)\
                        .set_shade_in_3d(True)
            tip = b_ghost.get_tip().set_opacity(opacity)
            self.add(b_ghost)
            anims += [UpdateFromFunc(b_ghost, update)]
        
        k_arrow = Arrow(start=[0,0,0], end=[2*self.amplitude,0,0], buff=0, tip_length=self.amplitude/10, stroke_color=WHITE)
        self.add(k_arrow)
                
        text_e = TexMobject(r'\boldsymbol{E}', color=BLUE_C)\
                    .rotate(90*DEGREES, RIGHT).rotate(90*DEGREES, OUT)\
                    .move_to(2 * UP + 0.5 * IN)
        text_b = TexMobject(r'\boldsymbol{B}', color=RED_C)\
                    .rotate(90*DEGREES, RIGHT).rotate(90*DEGREES, OUT)\
                    .move_to(2 * OUT + 0.5 * DOWN)
        text_k = TexMobject(r'\boldsymbol{k}', color=WHITE)\
                    .rotate(90*DEGREES, RIGHT).rotate(90*DEGREES, OUT)\
                    .move_to(2.5 * RIGHT + 0.5 * UP)
        self.add(text_e, text_b, text_k)
        
        anims += [UpdateFromFunc(text_e, lambda x: x.next_to(e_arrow.get_end(), 0.5*UP))]
        anims += [UpdateFromFunc(text_b, lambda x: x.next_to(b_arrow.get_end(), 0.5*DOWN))]
        
        self.play(*anims, run_time=OSC_RUNTIME, rate_func=linear)
```

```python
%%manim Polarization -i --media_dir ./med --video_output_dir ./vid --vcodec prores --verbose
# --vcodec prores

OSC_RUNTIME = 5
NTURNS = 2

class Polarization(Scene):
    CONFIG = {
        "amplitude": 3.5
    }
    def construct(self):
        import numpy as np
        self.camera.set_frame_width(8)
        self.camera.set_frame_center([0, 0, 0])
        self.camera.reset_pixel_shape(700, 700)
        self.camera.resize_frame_shape()
        
        time_tracker = ValueTracker(0)
        phase_tracker = ValueTracker(0)
        anims = [ApplyMethod(time_tracker.increment_value, OSC_RUNTIME)]
        anims += [UpdateFromFunc(phase_tracker, 
                                 lambda x: x.set_value(2 * PI * ((time_tracker.get_value()) / (OSC_RUNTIME / NTURNS))))]
                
        e0_arrow = Arrow(start=ORIGIN, end=ORIGIN+self.amplitude*UP, buff=0, tip_length=self.amplitude/20, stroke_color=BLUE_D)
        eL_arrow = Arrow(start=ORIGIN, end=ORIGIN+self.amplitude*UP/2, buff=0, tip_length=self.amplitude/20, stroke_color=TEAL_C)
        eR_arrow = Arrow(start=ORIGIN, end=ORIGIN+self.amplitude*UP/2, buff=0, tip_length=self.amplitude/20, stroke_color=PURPLE)
        eL_line = DashedLine(start=eL_arrow.get_end(), end=e0_arrow.get_end(), buff=0, stroke_color=WHITE, stroke_width=1)
        eR_line = DashedLine(start=eR_arrow.get_end(), end=e0_arrow.get_end(), buff=0, stroke_color=WHITE, stroke_width=1)
        e0_text = TexMobject(r'\boldsymbol{E}', color=BLUE_E)
        eL_text = TexMobject(r'\boldsymbol{E_L}', color=TEAL_E)
        eR_text = TexMobject(r'\boldsymbol{E_R}', color=PURPLE_E)
        self.add(e0_arrow, eL_arrow, eR_arrow)
        self.add(eL_line, eR_line)
        self.add(e0_text, eL_text, eR_text)
        def update_E0(obj):
            length = self.amplitude * np.cos(phase_tracker.get_value())
            obj[0].put_start_and_end_on(ORIGIN, ORIGIN + np.array([0, length, 0]))
            obj[1].next_to(obj[0].get_end(), RIGHT)
        def update_EL(obj):
            x = self.amplitude * np.sin(-phase_tracker.get_value())/2
            y = self.amplitude * np.cos(phase_tracker.get_value())/2
            obj[0].put_start_and_end_on(ORIGIN, ORIGIN + np.array([x, y, 0]))
            obj[1].next_to(obj[0].get_end(), LEFT)
            obj[2].put_start_and_end_on(obj[0].get_end(), e0_arrow.get_end())
        def update_ER(obj):
            x = self.amplitude * np.sin(phase_tracker.get_value())/2
            y = self.amplitude * np.cos(phase_tracker.get_value())/2
            obj[0].put_start_and_end_on(ORIGIN, ORIGIN + np.array([x, y, 0]))
            obj[1].next_to(obj[0].get_end(), RIGHT)
            obj[2].put_start_and_end_on(obj[0].get_end(), e0_arrow.get_end())
            
        anims += [UpdateFromFunc(VGroup(e0_arrow, e0_text), update_E0)]
        anims += [UpdateFromFunc(VGroup(eL_arrow, eL_text, eL_line), update_EL)]
        anims += [UpdateFromFunc(VGroup(eR_arrow, eR_text, eR_line), update_ER)]
        
        self.play(*anims, run_time=OSC_RUNTIME, rate_func=linear)
```

```python
%%manim Addition -i --media_dir ./med --video_output_dir ./vid --vcodec prores --verbose

OSC_RUNTIME = 5
NTURNS = 2

class Addition(ThreeDScene):
    CONFIG = {
        "camera_config": {
            "shading_factor": 1.0,
            "phi": 80 * DEGREES,
            "theta": 5 * DEGREES
        },
        "should_apply_shading": True,
        "color": False,
        "amplitude": 2
    }
    def construct(self):
        import numpy as np
        self.camera.set_frame_width(10)
        self.camera.set_frame_center([0, 0.25, -1.5])
        self.camera.reset_pixel_shape(800, 1000)
        self.camera.resize_frame_shape()
        
        time_tracker = ValueTracker(0)
        phase_tracker = ValueTracker(0)
        anims = [ApplyMethod(time_tracker.increment_value, OSC_RUNTIME)]
        anims += [UpdateFromFunc(phase_tracker, 
                                 lambda x: x.set_value(2 * PI * ((time_tracker.get_value()) / (OSC_RUNTIME / NTURNS))))]
                        
        d_angle = 2 * PI * NTURNS / (OSC_RUNTIME * self.camera.frame_rate)
        
        anims += self.addEgroup(shift=2.5*DOWN, angle=d_angle, col=PURPLE)
        anims += self.addEgroup(shift=2.5*UP, angle=-d_angle, col=TEAL_C)
        anims += self.addE0group(phase_tracker)
        
        plus = TexMobject('+')\
                    .rotate(90*DEGREES, RIGHT)\
                    .rotate(90*DEGREES, OUT)\
                    .move_to(0.5*OUT)
        equals = TexMobject('=').rotate(90*DEGREES, RIGHT)\
                    .rotate(90*DEGREES, OUT)\
                    .move_to(2*IN)
        self.add(plus, equals)
        
        self.play(*anims, run_time=OSC_RUNTIME, rate_func=linear)
    
    def addEgroup(self, shift, angle, col):
        import numpy as np
        def update_E(obj):
            obj.rotate(angle, LEFT, about_point=obj.get_start())

        e_arrow = Arrow(start=ORIGIN+shift, end=ORIGIN+shift+self.amplitude*UP/2, 
                        buff=0, tip_length=self.amplitude/10, stroke_color=col)\
                    .rotate(90*DEGREES, UP)
        self.add(e_arrow)
        anims = [UpdateFromFunc(e_arrow, update_E)]
        
        Nghosts = 20
        for i in range(Nghosts):
            opacity = ((Nghosts-1-i)/Nghosts)**2
            shift_x = -(i+1)/2
            
            y = self.amplitude * np.cos(shift_x)/2
            z = self.amplitude * np.sin(shift_x)/2
            
            origin = ORIGIN + shift + np.array([shift_x,0,0])
            e_ghost = Arrow(start=origin, end=origin+self.amplitude*UP/2, buff=0, tip_length=self.amplitude/10, 
                            stroke_color=col, stroke_opacity=opacity)\
                        .rotate(90*DEGREES, UP)\
                        .rotate(shift_x/2, LEFT, about_edge=DOWN)\
                        .set_shade_in_3d(True)
            tip = e_ghost.get_tip().set_opacity(opacity)
            self.add(e_ghost)
            anims += [UpdateFromFunc(e_ghost, update_E)]
        return anims
    
    def addE0group(self, phase):
        import numpy as np
        def update_E(obj):
            shift_x = obj.get_start()[0]
            delay = shift_x
            length = self.amplitude * np.cos(phase.get_value() + delay)
            new_origin = obj.get_start()
            obj.put_start_and_end_on(new_origin, new_origin + np.array([0, length, 0]))
            obj.rotate(90 * DEGREES, DOWN)
        origin = ORIGIN + 4.5*IN
        e_arrow = Arrow(start=origin, end=origin+np.array([0,self.amplitude,0]), buff=0, tip_length=self.amplitude/10, stroke_color=BLUE)
        self.add(e_arrow)
        anims = [UpdateFromFunc(e_arrow, update_E)]
        
        Nghosts = 20
        for i in range(Nghosts):
            opacity = ((Nghosts-1-i)/Nghosts)**2
            shift_x = -(i+3)/2+1
            new_origin = origin + np.array([shift_x, 0, 0])
            e_ghost = Arrow(start=new_origin, end=new_origin+np.array([0,self.amplitude,0]), buff=0, 
                            tip_length=self.amplitude/10, stroke_color=BLUE_E, stroke_opacity=opacity)\
                        .set_shade_in_3d(True)
            tip = e_ghost.get_tip().set_opacity(opacity)
            self.add(e_ghost)
            anims += [UpdateFromFunc(e_ghost, update_E)]
        return anims
```

```python
%%manim Answer -li --media_dir ./med --video_output_dir ./vid --vcodec prores --verbose

OSC_RUNTIME = 15*4
NTURNS = 20

class Answer(ThreeDScene):
    CONFIG = {
        "camera_config": {
            "shading_factor": 1.0,
            "phi": 70 * DEGREES,
            "theta": 5 * DEGREES,
            "frame_rate": 15
        },
        "fps" : 15,
        "should_apply_shading": True,
        "color": False,
        "amplitude": 2
    }
    def construct(self):
        import numpy as np
        self.camera.set_frame_width(10)
        self.camera.set_frame_center([0, 0.25, -1.5])
        self.camera.reset_pixel_shape(750, 900)
        self.camera.resize_frame_shape()
                
        time_tracker = ValueTracker(0)
        phase_tracker = ValueTracker(0)
        anims = [ApplyMethod(time_tracker.increment_value, OSC_RUNTIME)]
        anims += [UpdateFromFunc(phase_tracker, 
                                 lambda x: x.set_value(2 * PI * ((time_tracker.get_value()) / (OSC_RUNTIME / NTURNS))))]
                        
        d_angle = 2 * PI * NTURNS / (OSC_RUNTIME * self.camera.frame_rate)
        
        anim, e1, e1_g = self.addEgroup(shift=2.5*DOWN, angle=0.95*d_angle, col=PURPLE)
        anims += anim
        anim, e2, e2_g = self.addEgroup(shift=2.5*UP, angle=-d_angle, col=TEAL_C)
        anims += anim
        
        anims += self.addAnswer(e1, e1_g, e2, e2_g, BLUE_E)
                
        plus = TexMobject('+')\
                    .rotate(90*DEGREES, RIGHT)\
                    .rotate(90*DEGREES, OUT)\
                    .move_to(0.5*OUT)
        equals = TexMobject('=').rotate(90*DEGREES, RIGHT)\
                    .rotate(90*DEGREES, OUT)\
                    .move_to(2*IN)
        self.add(plus, equals)
        
        B_vector1 = Arrow(start=ORIGIN+2.5*DOWN+4*LEFT+1.5*IN, end=ORIGIN+2.5*DOWN+1.5*IN, buff=0, 
                         tip_length=self.amplitude/10, stroke_color=RED)\
                    .set_shade_in_3d(True)
        B_vector2 = Arrow(start=ORIGIN+2.5*UP+4*LEFT+1.5*IN, end=ORIGIN+2.5*UP+1.5*IN, buff=0, 
                         tip_length=self.amplitude/10, stroke_color=RED)\
                    .set_shade_in_3d(True)
        B_text1 = TexMobject(r'\boldsymbol{B}', color=RED).rotate(90*DEGREES, RIGHT)\
                    .rotate(90*DEGREES, OUT)\
                    .move_to(ORIGIN+2.5*DOWN+2*IN)
        B_text2 = TexMobject(r'\boldsymbol{B}', color=RED).rotate(90*DEGREES, RIGHT)\
                    .rotate(90*DEGREES, OUT)\
                    .move_to(ORIGIN+2.5*UP+2*IN)
        self.add(B_vector1, B_vector2)
        self.add(B_text1, B_text2)
        
        self.play(*anims, run_time=OSC_RUNTIME, rate_func=linear)
    
    def addAnswer(self, e1, e1_g, e2, e2_g, col):
        import numpy as np
        shift = 4 * IN
        def update_E(obj):
            obj.start = ORIGIN+shift
            obj.end = ORIGIN + shift + (e1.get_end() - e1.get_start()) + (e2.get_end() - e2.get_start())
            obj.generate_points()
            
        e_ans = Line(start=ORIGIN+shift, end=ORIGIN+shift+self.amplitude*LEFT, 
                        buff=0, tip_length=self.amplitude/10, stroke_color=col).set_shade_in_3d(True)
        self.add(e_ans)
        
        anims = [UpdateFromFunc(e_ans, update_E)]
        
        Nghosts = 20
        e_ghosts = []
        for i in range(Nghosts):
            opacity = ((Nghosts-1-i)/Nghosts)**2
            shift_x = -(i+1)/2
            origin = ORIGIN + shift + np.array([shift_x,0,0])
            e_ghost = Line(start=origin, end=origin+self.amplitude*LEFT, buff=0, 
                            stroke_color=col, stroke_opacity=opacity)\
                        .set_shade_in_3d(True)
            self.add(e_ghost)
            e_ghosts.append(e_ghost)
        
        def update_Eg(objs):
            for i, obj in enumerate(objs):
                shift_x = -(i+1)/2
                origin = ORIGIN + shift + np.array([shift_x,0,0])
                objs[i].start = origin
                objs[i].end = origin + (e1_g[i].get_end() - e1_g[i].get_start()) + (e2_g[i].get_end() - e2_g[i].get_start())
                objs[i].generate_points()

        anims += [UpdateFromFunc(VGroup(*e_ghosts), update_Eg)]
        
        return anims
    
    def addEgroup(self, shift, angle, col):
        import numpy as np
        def update_E(obj):
            obj.rotate(angle, LEFT, about_point=obj.get_start())
            
        e_arrow = Arrow(start=ORIGIN+shift, end=ORIGIN+shift+self.amplitude*UP/2, 
                        buff=0, tip_length=self.amplitude/10, stroke_color=col)\
                    .rotate(90*DEGREES, UP).set_shade_in_3d(True)
        self.add(e_arrow)
        anims = [UpdateFromFunc(e_arrow, update_E)]
        
        Nghosts = 20
        e_ghosts = []
        for i in range(Nghosts):
            opacity = ((Nghosts-1-i)/Nghosts)**2
            shift_x = -(i+1)/2
            
            y = self.amplitude * np.cos(shift_x)/2
            z = self.amplitude * np.sin(shift_x)/2
            
            origin = ORIGIN + shift + np.array([shift_x,0,0])
            e_ghost = Arrow(start=origin, end=origin+self.amplitude*UP/2, buff=0, tip_length=self.amplitude/10, 
                            stroke_color=col, stroke_opacity=opacity)\
                        .rotate(90*DEGREES, UP)\
                        .rotate(shift_x/2, LEFT, about_edge=DOWN)\
                        .set_shade_in_3d(True)
            tip = e_ghost.get_tip().set_opacity(opacity)
            self.add(e_ghost)
            e_ghosts.append(e_ghost)
            anims += [UpdateFromFunc(e_ghost, update_E)]
        return (anims, e_arrow, e_ghosts)
```
