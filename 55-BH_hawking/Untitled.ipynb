```python
import numpy as np
import os
os.environ['PATH'] = os.environ['PATH'] + ':/Library/TeX/texbin'
import matplotlib.pyplot as plt
import matplotlib
plt.style.use('ggplot')
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Georgia'],'size':25})
rc('text', usetex=True)
# rc('text.latex',unicode=True)
rc('text.latex',preamble='\\usepackage[utf8]{inputenc}')
rc('text.latex',preamble='\\usepackage[russian]{babel}')
# rc('text.latex',preamble='\\usepackage{mathabx}')

import matplotlib.ticker as mticker

class MathTextSciFormatter(mticker.Formatter):
  def __init__(self, fmt="%1.2e"):
    self.fmt = fmt
  def __call__(self, x, pos=None):
    s = self.fmt % x
    decimal_point = '.'
    positive_sign = '+'
    tup = s.split('e')
    significand = tup[0].rstrip(decimal_point)
    sign = tup[1][0].replace(positive_sign, '')
    exponent = tup[1][1:].lstrip('0')
    if exponent:
      exponent = '10^{%s%s}' % (sign, exponent)
    if significand and exponent:
      s =  r'%s{\times}%s' % (significand, exponent)
    else:
      s =  r'%s%s' % (significand, exponent)
    return "${}$".format(s)
```

```python
xs = np.logspace(-5, 34, 50)
ys = (16/3) * 2.658e-38 * (xs)**3
def YrToS(x):
  return x * (365 * 24 * 3600)
def SToYr(x):
  return x / (365 * 24 * 3600)

fig = plt.figure(figsize=(15, 15))
ax1 = plt.subplot(311)
ax2 = plt.subplot(312)
ax3 = plt.subplot(313)

ax1.plot(xs, ys)
ax2.plot(xs, ys)
ax3.plot(xs, ys)

ax1.set_xlim(xs.min(), xs.max())
ax2.set_xlim(xs.min(), xs.max())
ax3.set_xlim(xs.min(), xs.max())
ax1.set_ylim(1e32, 1e70);
ax2.set_ylim(1e-15, 1e25);
ax3.set_ylim(1e-46, 1e-20);
ax1.set_xscale('log'); ax1.set_yscale('log');
ax2.set_xscale('log'); ax2.set_yscale('log');
ax3.set_xscale('log'); ax3.set_yscale('log');

ax2.set_ylabel(r'время испарения, с');

secaxy1 = ax1.secondary_yaxis('right', functions=(SToYr, YrToS))
secaxy1.set_yticks([1e34]);
ax1.axhline(1e34 * (365 * 24 * 3600), c='lightgray', ls='--', alpha=0.6)
secaxy1.set_yticklabels(['период полураспада\nпротона (нижняя оценка)']);

secaxy2 = ax2.secondary_yaxis('right', functions=(SToYr, YrToS))
secaxy2.set_yticks([1e-12 / (365 * 24 * 3600), 1 / (365 * 24 * 3600), 1, 14e9]);
ax2.axhline(1e-12, c='lightgray', ls='--', alpha=0.6)
ax2.axhline(1, c='lightgray', ls='--', alpha=0.6)
ax2.axhline((365 * 24 * 3600), c='lightgray', ls='--', alpha=0.6)
ax2.axhline(14e9 * (365 * 24 * 3600), c='lightgray', ls='--', alpha=0.6)
secaxy2.set_yticklabels(['период полураспада\n$c$ и $b$ кварка', '1 с', '1 год', 'возраст Вселенной']);

secaxy3 = ax3.secondary_yaxis('right', functions=(SToYr, YrToS))
secaxy3.set_yticks([1e-25 / (365 * 24 * 3600), 1e-43 / (365 * 24 * 3600)]);
ax3.axhline(1e-43, c='lightgray', ls='--', alpha=0.6)
ax3.axhline(1e-25, c='lightgray', ls='--', alpha=0.6)
secaxy3.set_yticklabels(['период полураспада\n$W^\pm/Z^0$ бозонов', 'планковское время']);

ax1.spines['bottom'].set_visible(False)
ax2.spines['bottom'].set_visible(False)
ax3.spines['top'].set_visible(False)
ax1.tick_params(axis='x', which='both', bottom=False, labelbottom=False)
ax2.tick_params(axis='x', which='both', bottom=False, labelbottom=False)

d = .01
kwargs = dict(transform=ax1.transAxes, color='k', clip_on=False)
ax1.plot((-d, +d), (-d, +d), **kwargs)
ax1.plot((1 - d, 1 + d), (-d, +d), **kwargs)

kwargs.update(transform=ax2.transAxes)
ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)
ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs);

kwargs = dict(transform=ax2.transAxes, color='k', clip_on=False)
ax2.plot((-d, +d), (-d, +d), **kwargs)
ax2.plot((1 - d, 1 + d), (-d, +d), **kwargs)

kwargs.update(transform=ax3.transAxes)
ax3.plot((-d, +d), (1 - d, 1 + d), **kwargs)
ax3.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs);

ax3.set_xlabel('масса, г');

secaxx1 = ax1.secondary_xaxis('top')
secaxx1.set_xticks([50*1e3, 1e6*1e6, 0.022 * 6e27 / 100, 6e27, 2e33], minor=False);
ax1.axvline(50*1e3, c='lightgray', ls='--', alpha=0.6)
ax2.axvline(50*1e3, c='lightgray', ls='--', alpha=0.6)
ax3.axvline(50*1e3, c='lightgray', ls='--', alpha=0.6)

ax1.axvline(1e6*1e6, c='lightgray', ls='--', alpha=0.6)
ax2.axvline(1e6*1e6, c='lightgray', ls='--', alpha=0.6)
ax3.axvline(1e6*1e6, c='lightgray', ls='--', alpha=0.6)

ax1.axvline(0.022 * 6e27 / 100, c='lightgray', ls='--', alpha=0.6)
ax2.axvline(0.022 * 6e27 / 100, c='lightgray', ls='--', alpha=0.6)
ax3.axvline(0.022 * 6e27 / 100, c='lightgray', ls='--', alpha=0.6)

ax1.axvline(6e27, c='lightgray', ls='--', alpha=0.6)
ax2.axvline(6e27, c='lightgray', ls='--', alpha=0.6)
ax3.axvline(6e27, c='lightgray', ls='--', alpha=0.6)

ax1.axvline(2e33, c='lightgray', ls='--', alpha=0.6)
ax2.axvline(2e33, c='lightgray', ls='--', alpha=0.6)
ax3.axvline(2e33, c='lightgray', ls='--', alpha=0.6)

secaxx1.set_xticklabels(['человек', 'небоскрёб', 'мировой\nокеан', r'$\bigoplus$', r'$\bigodot$']);

# secaxx3 = ax3.secondary_xaxis('bottom')

# def make_patch_spines_invisible(ax):
#   ax.set_frame_on(True)
#   ax.patch.set_visible(False)
#   for sp in ax.spines.values():
#     sp.set_visible(False)
# secaxx3 = ax3.twiny()
# secaxx3.xaxis.set_ticks_position('bottom') # set the position of the second x-axis to bottom
# secaxx3.xaxis.set_label_position('bottom') # set the position of the second x-axis to bottom
# secaxx3.spines['bottom'].set_position(('outward', 80))
# secaxx3.set_xlabel('температура Хокинга, эВ')

# def Grams2THinK(x):
#   return 1e28 / x

# def THinK2Grams(x):
#   return x * 1e28

# secaxx3.set_xscale('log')
# secaxx3.set_xlim(xs.min(), xs.max())
# GeVs = 10.0**np.arange(1, 120, 25)
# secaxx3.set_xticks(THinK2Grams(GeVs * 8.6e-5 / 1e9))
# # secaxx3.xaxis.set_major_formatter(MathTextSciFormatter("%1.2e"))
# secaxx3.set_xticklabels(GeVs[::-1], fmt=MathTextSciFormatter("%1.2e"))
# secaxx3.xaxis.grid(False)
# # secaxx3.set_xlim(Grams2THinK(xs.min()), Grams2THinK(xs.max()))
# # secaxx3.set_xticks([])

# secaxx3.spines["bottom"].set_position(("axes", 1.2))
# make_patch_spines_invisible(secaxx3)
# secaxx3.spines["bottom"].set_visible(True)

# secaxx11.set_xticks([50*1e3, 1e6*1e6, 0.022 * 6e27 / 100, 6e27, 2e33], minor=False);

plt.subplots_adjust(hspace = .1)
plt.tight_layout()
plt.savefig("evaporation.png", dpi=100)
plt.close()
```
